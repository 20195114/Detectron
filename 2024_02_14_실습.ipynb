```python
# 데이터 시각화
#데이터 시각화의 필요성
#앤스콤데이터 가져오기
import seaborn as sns
anscombe = sns.load_dataset("anscombe")
print(anscombe.head())
print(anscombe['dataset'].unique())
```

```python
#dataset 별로 분리
dataset_1 = anscombe[anscombe['dataset'] == '|']
dataset_2 = anscombe[anscombe['dataset'] == '||']
dataset_3 = anscombe[anscombe['dataset'] == '|||']
dataset_4 = anscombe[anscombe['dataset'] == '|V']
```

```python
print(dataset_1)
```

```python
print(dataset_1['x'].std())
print(dataset_2['x'].std())
print(dataset_3['x'].std())
print(dataset_4['x'].std())
```

```python
# 데이터 시각화
import matplotlib.pyplot as plt
fig = plt.figure()

axes1 = fig.add_subplot(2, 2, 1)
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)

axes1.plot(dataset_1['x'], dataset_1['y'], 'o')
axes2.plot(dataset_2['x'], dataset_2['y'], 'o')
axes3.plot(dataset_3['x'], dataset_3['y'], 'o')
axes4.plot(dataset_4['x'], dataset_4['y'], 'o')

axes1.set_title("dataset_1")
axes2.set_title("dataset_2")
axes3.set_title("dataset_3")
axes4.set_title("dataset_4")
```

```python
# 라인 그래프
# 엑셀 파일 읽어오기
# 첫 행이 컬럼이름
df = pd.read_excel('/content/시도_별_이동자수.xlsx', header=0)
print(df.head())
#엑셀에서 셀병합이 있으면 첫번째를 제외하고는 NaN으로 처리
# NaN 데이터를 앞의 데이터로 채우기
df = df.fillna(method = 'ffill')
```

```python
#전출지별이 서울특별시이고 접입지별이 서울특별시가 아닌 데이터만 추출
#조건 생성
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
print(df_seoul.head())
```

```python
#컬럼 제거
#df_seoul.drop(['전출지별'], axis = 1, inplace = True)
#전입지별이라를 컬럼이름을 전입지로 변경
df_seoul.rename({'전입지별':'전입지'}, axis = 1, inplace = True)
#전입지를 인덱스로 설정
df_seoul.set_index('전입지', inplace = True)
print(df_seoul.head())
```

```python
sr_one = df_seoul.loc['전라남도']
print(sr_one)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 산포도: 데이터의 분포나 상관관계를 파악하기 위해서 그립니다
df = pd.read_csv('auto-mpg.csv')
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
              'acceleration', 'model year', 'origin', 'name']

size = df['cylinders'] / df['cylinders'].max() * 200
plt.scatter(x=df['weight'], y=df['mpg'], s=size, c='coral', alpha=0.5)
plt.show()
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
# 산점도 와 회귀선 출력
plt.figure(figsize = (8, 6))

sns.regplot(x = "total_bill", y = "tip", data = tips)

plt.show()
```

```python
# 산점도 와 회귀선 출력
plt.figure(figsize = (8, 6)) #캔버스 크기 설정

sns.lmplot(x = "total_bill", y = "tip",hue = 'smoker' ,data = tips)

plt.show()
```

```python
flights = sns.load_dataset("flights")
#index를 month 로 column 에 year를 passenger의 합계를 구하는 피벗을 생성
#values는 여러개 설정이 가능
data = flights.pivot(index = "month", columns = "year", values = "passengers")
#print(data)

#히트맵을 출력하는데 값을 레이블로 설정하고 포맷을 정수
sns.heatmap(data,annot = True, fmt = 'd')
plt.show()
```

```python
# 데이터의 밀집 형태도 확인이 가능
tips = sns.load_dataset('tips')
#print(tips)
sns.reset_defaults()
sns.boxplot(x="day", y = "total_bill", data= tips)
sns.swarmplot(x="day", y = "total_bill", data= tips)
plt.show()
```

```python
sns.pairplot(tips[["total_bill", "tip", "size"]])
plt.show()
```

```python
import plotnine

df = pd.DataFrame({
    'letter' : ['Alpha', 'Beta', 'Delta', 'Gamma'] * 2,
    'pos' : [1, 2, 3, 4] * 2,
    'num_of_letters' : [5, 4, 5, 5] * 2
})

(plotnine.ggplot(df)
+plotnine.geom_col(plotnine.aes(x='letter', y='pos', color='letter'))
+plotnine.geom_line(plotnine.aes(x='letter', y='num_of_letters', color='letter'), size=1)
+plotnine.scale_color_hue(l = 0.45)
+plotnine.ggtitle('Greek Letter Analysis')
)
```
